# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline 

# 
QUERY = """
SELECT *
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

count, bin_edges = np.histogram(df['ConvertedComp'].dropna())

df['ConvertedComp'].plot(kind='hist', figsize=(8, 5), xticks=bin_edges)

plt.title('Histogram of Salary converted to annual USD salaries') # add a title to the histogram
plt.ylabel('Salary in USD') # add y-label
plt.xlabel('Number of Salary') # add x-label

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

df['Age'].plot(kind='box', figsize=(20,10))

plt.title('Box plot of Salalry in USD')
plt.ylabel('Number of Immigrants')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
df.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10, 6), color='darkblue')

plt.title('WorkWeekHrs by Age')
plt.xlabel('Age')
plt.ylabel('Number of WorkWeekHrs')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYear, count(*)
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by count(*) desc
limit 5

"""
database=pd.read_sql_query(QUERY,conn)

#plot
database["count(*)"].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90,     # start angle 90°
                            labels=None,)


plt.title('Top 5 databases to learn next year')
plt.axis('equal') # Sets the pie chart to look like a circle.

# add legend
plt.legend(labels=database["DatabaseDesireNextYear"], loc='upper left')

plt.show()
```

### Stacked Charts


WorkWeekHrsCreate a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age, sum(WorkWeekHrs)/count(WorkWeekHrs) as WorkWeekHrs, sum(CodeRevHrs)/count(CodeRevHrs) as CodeRevHrs
FROM master
where Age between 30 and 35
group by Age

"""
data=pd.read_sql_query(QUERY,conn)

#Plot
width = 0.5     
fig, ax = plt.subplots()

ax.bar(data["Age"], data["WorkWeekHrs"], width, label='WorkWeekHrs')
ax.bar(data["Age"], data["CodeRevHrs"], width, bottom=data["WorkWeekHrs"],
       label='CodeRevHrs')

ax.set_ylabel('Hours')
ax.set_title('Hours for the age group 30 to 35')

plt.legend(loc='lower right')

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master
where Age between 45 and 60
order by Age
"""
data=pd.read_sql_query(QUERY,conn)
data['ConvertedComp'].median()
data['Median'] = data.groupby('Age')['ConvertedComp'].transform('median')
data_final=data[["Age",'Median']].drop_duplicates()
data_final

#Plot
plt.plot(data_final['Age'], data_final['Median'])
plt.title('ConvertedComp for all ages from 45 to 60')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
plt.show()

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch,COUNT(*) as count
FROM master
group by MainBranch
order by MainBranch
"""
data=pd.read_sql_query(QUERY,conn)
#plot

plt.barh(data['count'], data['MainBranch'])

plt.show()
```
